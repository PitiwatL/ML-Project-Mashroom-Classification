```python
# 23 Features
f = open('Dataset\\mushroom\\agaricus-lepiota.data', 'r')
f = f.read()

clss = ['p', 'e']
Class = []
missing = []
count = [[], []]
for idx in tqdm(range(len(f.split('\n')))):
    try:
        Class.append(f.split('\n')[idx].split(',')[0])
        idx = clss.index(f.split('\n')[idx].split(',')[0])
        count[idx].append(f.split('\n')[idx].split(',')[0])
        
    except: 
        missing.append(idx)
```
